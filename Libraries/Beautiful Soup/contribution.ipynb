```python
from bs4 import BeautifulSoup
import requests
import matplotlib.pyplot as plt
import numpy as np

user = 'ayu4tor'
month = 2
gurl = f'https://github.com/{user}'

r = requests.get(gurl)
soup = BeautifulSoup(r.content, 'html5lib')

a = soup.findAll('rect', attrs = {'class':'ContributionCalendar-day'})
x, y = [], []

for i in a:
    try:
        c = int(i['data-count'])
        d = i['data-date']
        y.append(c)
        x.append(d)
    except:
        pass
    
x = x[::-1][:30*month][::-1]
y = y[::-1][:30*month][::-1]
fig = plt.figure(figsize=(10*month,3*month))

xpoints = np.array(x)
ypoints = np.array(y)
plt.xticks(rotation=90)
plt.grid()

plt.title(f'GitHub username : {user}')
plt.xlabel(f"Date (of last {month} month(s))")
plt.ylabel("Git Commit")

plt.plot(xpoints, ypoints)
plt.show()
fig.savefig(f'user/{user}.png')
```

```python
from bs4 import BeautifulSoup
import requests

url = 'https://inshorts.com/en/read/'
r = requests.get(url)
soup = BeautifulSoup(r.content, 'html5lib')

heading = soup.findAll('span', attrs = {'itemprop':'headline'})
imagex = soup.findAll('div', attrs = {'class':'news-card-image'})
bodyx = soup.findAll('div', attrs = {'itemprop':'articleBody'})
readmore = soup.findAll('div', attrs = {'class':'read-more'})

for i in range(len(heading)):
    print(i+1, '). Title...')
    print('\t', heading[i].text)
    
    print('\nImage link...')
    print('\t', imagex[i]['style'].split("'")[1])
    
    print('\n60 words news...')
    print('\t', bodyx[i].text)
    
    try:
        print('\nRead more...')
        print('\t', readmore[i].a['href'])
    except:
        print('Link not found !')
    
    print('\n', '-'*120, end='\n\n')
```

```python
# !pip install bs4

from bs4 import BeautifulSoup
import requests

user = 'CleuJunior'
gurl = f'https://github.com/{user}'

r = requests.get(gurl)
soup = BeautifulSoup(r.content, 'html5lib')

a = soup.findAll('rect', attrs = {'class':'ContributionCalendar-day'})
x = []
y = []

for i in a:
    try:
        x.append(i['data-date'])
        y.append(int(i['data-count']))
    except:
        pass
    
x1 = x[::-1][:10][::-1]
y1 = y[::-1][:10][::-1]

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array(x1)
ypoints = np.array(y1)

plt.xticks(rotation=90)
plt.grid()

plt.plot(xpoints, ypoints)
plt.show()
```
